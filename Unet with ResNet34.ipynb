```python
import os
import random
import numpy as np
import time
from PIL import Image
import cv2
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
from torchvision import transforms
from pytorch_msssim import ssim
import segmentation_models_pytorch as smp
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
from tqdm import tqdm

# --------- Reproducibility ---------
torch.manual_seed(42)
random.seed(42)
np.random.seed(42)

# --------- Dataset ---------
class PairedDataset(Dataset):
    def __init__(self, foggy_dir, clear_dir, transform=None):
        self.foggy_imgs = sorted(os.listdir(foggy_dir))
        self.clear_imgs = sorted(os.listdir(clear_dir))
        self.foggy_dir = foggy_dir
        self.clear_dir = clear_dir
        self.transform = transform

    def __len__(self):
        return len(self.foggy_imgs)

    def __getitem__(self, idx):
        foggy = Image.open(os.path.join(self.foggy_dir, self.foggy_imgs[idx])).convert('RGB')
        clear = Image.open(os.path.join(self.clear_dir, self.clear_imgs[idx])).convert('RGB')
        if self.transform:
            foggy = self.transform(foggy)
            clear = self.transform(clear)
        return foggy, clear

# --------- Color Loss in Lab Space ---------
class ColorLossLab(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, pred, target):
        pred_np = pred.detach().cpu().numpy()
        target_np = target.detach().cpu().numpy()
        loss = 0
        for p, t in zip(pred_np, target_np):
            p_img = np.transpose(p, (1, 2, 0))
            t_img = np.transpose(t, (1, 2, 0))
            p_lab = cv2.cvtColor((p_img * 255).astype(np.uint8), cv2.COLOR_RGB2LAB)
            t_lab = cv2.cvtColor((t_img * 255).astype(np.uint8), cv2.COLOR_RGB2LAB)
            loss += np.mean(np.abs(p_lab[..., 1:] - t_lab[..., 1:])) / 255.0
        return torch.tensor(loss / len(pred), requires_grad=True).to(pred.device)

# --------- Composite Loss ---------
color_loss_fn = ColorLossLab()

def composite_loss(output, target, alpha=0.8, beta=0.15, gamma=0.05):
    mse = nn.functional.mse_loss(output, target)
    ssim_val = ssim(output, target, data_range=1.0, size_average=True)
    color = color_loss_fn(output, target)
    return alpha * mse + beta * (1 - ssim_val) + gamma * color

# --------- Evaluation Metrics ---------
def get_metrics(model, loader):
    model.eval()
    total_psnr, total_ssim = 0, 0
    with torch.no_grad():
        for foggy, clear in loader:
            foggy, clear = foggy.to(device), clear.to(device)
            output = torch.clamp(model(foggy), 0.0, 1.0)
            total_psnr += psnr(clear.cpu().numpy(), output.cpu().numpy(), data_range=1.0)
            total_ssim += ssim(output, clear, data_range=1.0, size_average=True).item()
    n = len(loader)
    return total_psnr / n, total_ssim / n

# --------- Config ---------
data_dir = './foggy_dataset/'
batch_size = 8
epochs = 500
lr = 1e-4
img_size = (256, 256)

# --------- Logging Setup ---------
log_file = open("training_log.txt", "w")

# --------- Transforms ---------
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.1, contrast=0.1),
    transforms.ToTensor()
])

# --------- DataLoader ---------
dataset = PairedDataset(
    foggy_dir=os.path.join(data_dir, 'foggy'),
    clear_dir=os.path.join(data_dir, 'clear'),
    transform=transform
)
val_size = int(0.1 * len(dataset))
train_size = len(dataset) - val_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False, num_workers=0, pin_memory=True)

# --------- Model ---------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = smp.Unet(
    encoder_name="resnet34",
    encoder_weights="imagenet",
    in_channels=3,
    classes=3,
    activation=None
).to(device)

# --------- Optimizer and Scheduler ---------
optimizer = optim.Adam(model.parameters(), lr=lr)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)

# --------- Training ---------
best_psnr = 0
for epoch in range(epochs):
    start_time = time.time()

    model.train()
    total_loss = 0
    for foggy, clear in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
        foggy, clear = foggy.to(device), clear.to(device)
        output = model(foggy)
        output = torch.clamp(output, 0.0, 1.0)

        loss = composite_loss(output, clear)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_psnr, avg_ssim = get_metrics(model, val_loader)
    scheduler.step()

    elapsed = time.time() - start_time
    log_line = f"[Epoch {epoch+1}/{epochs}] Time: {elapsed:.2f}s | Loss: {total_loss:.4f} | PSNR: {avg_psnr:.2f} dB | SSIM: {avg_ssim:.4f}"
    print(log_line)
    log_file.write(log_line + "\n")
    log_file.flush()

    if avg_psnr > best_psnr:
        best_psnr = avg_psnr
        torch.save(model.state_dict(), "best_resnet34_unet_colortry1.pth")
        print("✅ Saved new best model.")
        log_file.write("Saved new best model.\n")

log_file.close()

# --------- Contrast Stretching ---------
def contrast_stretch(img):
    p1, p99 = np.percentile(img, (1, 99))
    return np.clip((img - p1) / (p99 - p1 + 1e-5), 0, 1)

# --------- Show Prediction ---------
def show_prediction(idx=0):
    foggy, clear = val_dataset[idx]
    model.eval()
    with torch.no_grad():
        input_tensor = foggy.unsqueeze(0).to(device)
        output = model(input_tensor).squeeze(0).cpu()
        output = torch.clamp(output, 0.0, 1.0)

    def to_img(t): return np.clip(t.permute(1, 2, 0).numpy(), 0, 1)

    foggy_np = to_img(foggy)
    clear_np = to_img(clear)
    pred_np = contrast_stretch(to_img(output))

    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(foggy_np); axs[0].set_title("Foggy Input")
    axs[1].imshow(clear_np); axs[1].set_title("Ground Truth")
    axs[2].imshow(pred_np); axs[2].set_title("Predicted (Enhanced)")
    for ax in axs: ax.axis("off")
    plt.tight_layout(); plt.show()

# Visualize
show_prediction(idx=0)
```

```python
from PIL import Image
from torchvision import transforms

def predict_single_image(image_path):
    model.eval()
    
    # 🔧 Use the same preprocessing as training (NO flip, NO jitter)
    preprocess = transforms.Compose([
        transforms.Resize((256, 256)),     # Same as training
        transforms.ToTensor(),             # Converts to [0, 1] and CHW
    ])

    # Load image
    img = Image.open(image_path).convert("RGB")
    input_tensor = preprocess(img).unsqueeze(0).to(device)  # [1, 3, 256, 256]
    print("Input tensor shape:", input_tensor.shape)
    print("Min/max:", input_tensor.min().item(), input_tensor.max().item())

    # Model prediction
    with torch.no_grad():
        output = model(input_tensor).squeeze(0).cpu()
        output = torch.clamp(output, 0.0, 1.0)

    # Convert tensors to numpy for visualization
    def to_img(t): return np.clip(t.permute(1, 2, 0).numpy(), 0, 1)
    input_np = to_img(input_tensor.squeeze(0).cpu())
    pred_np = contrast_stretch(to_img(output))  # Optional enhancement

    # Visualize
    fig, axs = plt.subplots(1, 2, figsize=(10, 4))
    axs[0].imshow(input_np); axs[0].set_title("Input Foggy Image")
    axs[1].imshow(pred_np); axs[1].set_title("Predicted Clear (Enhanced)")
    for ax in axs: ax.axis("off")
    plt.tight_layout(); plt.show()


predict_single_image("./foggy citi/Medium_Fog/026.png")
```
